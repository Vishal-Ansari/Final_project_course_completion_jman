Joining the tables

```python

# Join course and modules on course_id
employee_enrollment = pd.merge(employee_df, enrollment_df, on='employee_id', how='inner')

# Join with course on course_id
employee_enrollment_course = pd.merge(employee_enrollment, course_df, on='course_id', how='inner')


# Join with performance on employee_id
final_result = pd.merge(employee_enrollment_course, performance_df, on='employee_id', how='inner')


# Save the final result to a CSV file
# final_result.to_csv('../../data/reporting_data/final_combined_data.csv', index=False)
```

```python
final_result.head()
```

Creating the Reporting table

```python
# Select relevant columns for the reporting table
reporting_columns = [
    'employee_id', 'name', 'department', 
    'course_id_x', 'title', 
    'completion_status', 'completion_percentage', 
    'average_score', 'time_spent_x'
]

# Create a reporting table
reporting_table = final_result[reporting_columns]

# Calculate average completion time per course
average_completion_time = final_result.groupby('course_id_x').agg(
    expected_time_to_complete=('time_spent_x', 'mean')
).reset_index()

# Merge the average completion time back into the reporting table
reporting_table = reporting_table.merge(average_completion_time, on='course_id_x', how='left')

# Calculate additional metrics
reporting_table = reporting_table.groupby(
    ['employee_id', 'name', 'department', 'course_id_x', 'title']
).agg(
    average_completion_percentage=('completion_percentage', 'mean'),
    average_score=('average_score', 'mean'),
    average_time_to_completion=('time_spent_x', 'mean'),  # Average time taken per course
    expected_time_to_complete=('expected_time_to_complete', 'mean')  # Adding expected time to complete
).reset_index()

# Rename columns for clarity
reporting_table.rename(columns={
    'average_completion_percentage': 'Average Completion Percentage',
    'average_score': 'Average Score',
    'average_time_to_completion': 'Average Time to Completion',
    'expected_time_to_complete': 'Expected Time to Complete'
}, inplace=True)

# Save the reporting table to a CSV file
# reporting_table.to_csv('../../data/reporting_data/reporting_table.csv', index=False)

# Display the reporting table
reporting_table.head()
```

Creating the KPIs's

```python

# 1. Average Completion Rate
avg_completion_rate = reporting_table.groupby('employee_id').agg(
    Average_Completion_Rate=('Average Completion Percentage', 'mean')
).reset_index()

# 2. Average Score per Course
avg_score_per_course = reporting_table.groupby('course_id_x').agg(
    Average_Score=('Average Score', 'mean')
).reset_index()

# 3. Completion Time Efficiency
reporting_table['Time_Efficiency'] = (reporting_table['Average Time to Completion'] / reporting_table['Expected Time to Complete'])*100
completion_time_efficiency = reporting_table.groupby('course_id_x').agg(
    Avg_Time_Efficiency=('Time_Efficiency', 'mean')
).reset_index()


# 4. Departmental Performance
departmental_performance = reporting_table.groupby('department').agg(
    Avg_Completion_Rate=('Average Completion Percentage', 'mean'),
    Avg_Score=('Average Score', 'mean'),
    Avg_Time_To_Completion=('Average Time to Completion', 'mean')
).reset_index()

# 5. Total Time Spent on Learning
total_time_spent = reporting_table.groupby('employee_id').agg(
    Total_Time_Spent=('Average Time to Completion', 'sum')
).reset_index()



# 6. High Performer Identification
high_performers = reporting_table.nlargest(10, 'Average Score')

# 7. Low Performer Analysis
low_performers = reporting_table.nsmallest(10, 'Average Score')

# Display the calculated KPIs
print("Average Completion Rate:\n", avg_completion_rate)
print("Average Score per Course:\n", avg_score_per_course)
print("Completion Time Efficiency:\n", completion_time_efficiency)
print("Departmental Performance:\n", departmental_performance)
print("Total Time Spent on Learning:\n", total_time_spent)
print("High Performers:\n", high_performers)
print("Low Performers:\n", low_performers)
```

Saving the Kpi's


```python
save_path = '../../data/reporting_data/kpis/'

avg_completion_rate.to_csv(f'{save_path}avg_completion_rate.csv', index=False)
avg_score_per_course.to_csv(f'{save_path}avg_score_per_course.csv', index=False)
completion_time_efficiency.to_csv(f'{save_path}completion_time_efficiency.csv', index=False)
departmental_performance.to_csv(f'{save_path}departmental_performance.csv', index=False)
total_time_spent.to_csv(f'{save_path}total_time_spent.csv', index=False)
high_performers.to_csv(f'{save_path}high_performers.csv', index=False)
low_performers.to_csv(f'{save_path}low_performers.csv', index=False)

print("Data saved to CSV files in ../../data/reporting_data/kpis/.")
```
